```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.plot(x,y,color='blue',linestyle='dashed',linewidth=2,marker='o')
plt.title("Customized Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[1000,2000,3000,4000,5000,6000]
plt.plot(x,y,color='green',linestyle='dashed',marker='o')
plt.title("Savings Line Plot")
plt.xlabel("Months")
plt.ylabel("Savings amount")
plt.show()
```

```python
cat=['A','B','C','D','E']
val=[10,15,7,12,9]
plt.bar(cat,val,color='skyblue')
plt.title("Basic Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
plt.bar(cat,val,color=['red','green','yellow','pink','purple'],edgecolor='black')
plt.show()
```

```python
plt.bar(cat,val,width=0.5)
```

```python
ages=[20,22,22,23,25,27,27,28,30,31,32,34,35,35,35,37,39,42,45,47,49,51,54,57,59,62,65]
b=[20,30,40,50,60,70]
plt.hist(ages,bins=b,color='skyblue',edgecolor='black')
plt.title("Age Distribution of Customers")
plt.xlabel("Age Ranges")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
x=['python','ruby','c++','java']
y=[20,30,40,50]
plt.pie(y,labels=x,colors=['lightgreen','yellow','pink','orange'], autopct='%1.1f%%',startangle=90)
plt.title("Programming Language Usage")
plt.legend()
```

```python
x=[5,10,15,20,25]
y=[10,20,25,40,35]
plt.scatter(x,y,color='blue',s=100,edgecolor='black')
plt.title("Scatter Plot Example")
plt.xlabel("X-values")
plt.ylabel("y=values")
plt.show()
```
